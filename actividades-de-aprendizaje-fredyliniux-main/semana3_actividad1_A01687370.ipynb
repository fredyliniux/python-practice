De acuerdo a la informacion de los datos, se van a susutituir los valores numericos por la media y la moda segun convenga para reemplazarlos por valores nulos o NaN

Creamos una lista con las columnas a las que les faltan valores y se completaran con la moda

````python
moda=['X2','X3','X4','X6','X7','X8','X9','X10','X11','Y']
````

````python
for i in moda:
  media=ndf[i].mode()
  ndf[i].fillna(value = media, 
                    inplace = True)
ndf
````

````python
ndf.isnull().any()
````

````python
ndf.isna().any()
````

Creamos una lista con las columnas a las que les faltan valores y se completaran con la media

````python
med=['X5','X12','X13','X14','X15','X16', 'X17', 'X18', 'X19','X20','X21','X22','X23']
````

Iteramos la lista y sustituimos los valores nulos con el valor de la media de la columna.

````python
for i in med:
  media=ndf[i].mean()
  ndf[i].fillna(value = media, 
                    inplace = True)
ndf
````
