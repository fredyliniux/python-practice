En esta sección se sustituirán los valores nulos por datos calculados (media/moda)

Se genera una lista con los nombres de las columnas donde se encontraron valores nulos y se completaran con la moda

```python
mod=['X2',	'X3',	'X4',	'X6',	'X7',	'X8',	'X9',	'X10',	'X11','Y']
```

Se hace una iteración de la lista para calcular la moda de cada columna y sustituya el valor resultante en donde encuentre valores nulos

```python
for i in mod:
  moda=nA[i].mode()
  moda[0]
  nA[i].fillna(value = moda[0], 
                    inplace = True)
nA
```

Se genera una lista con los nombres de las columanas donde se encontraron valores nulos y se completaran con la media

```python
med=['X5', 'X12',	'X13',	'X14',	'X15',	'X16',	'X17',	'X18',	'X19',	'X20',	'X21',	'X22',	'X23']
```

Se hace una iteración de la lista para calcular la media de cada columna y sustituya el valor resultante en donde encuentre valores nulos

```python
for i in med:
  media=nA[i].mean()
  nA[i].fillna(value = media, 
                    inplace = True)
```

Se comprueba que ya no haya valores nulos
