```
#@title Problema 3 Fibonaccis modulares
#@markdown Evaluar el  Fibonacci número n mod 10000 para 1<n<1000000000000000000000 

n =  1999998655643425132151#@param {type:"number"}
import time




def fibo(n):
  S=[1,1]
  if(n<2):
    return 1
  else:
    for i in range(n):
      g=len(S)-1
      S=S+[(S[g]+S[g-1])%10000]
    return S[g] 

start = time.time()
print(fibo(n%15000))
end = time.time()

print("tiempo=",end - start)
```

```
#@title Problema 4 Monedas
#@markdown En un  sistema monetario donde existen exclusivamente monedas de 1, 2, 5, 10 y 20 pesos, determinar la cantidad de formas en las que una cantidad dada de pesos puede ser expresada.

#@markdown -Por ejemplo: para n=5 se tiene:

#@markdown 5 =1 + 1 + 1 + 1 + 1 

#@markdown 5= 2 + 1 + 1 + 1

#@markdown 5=2 + 2 + 1

#@markdown 5= 5

#@markdown Entonces para 5 hay 4 maneras de darse con monedas.Encontrar esta cantidad para n <10000000

n =  500#@param {type:"number"}

    

def count(S, m, n): 

    table = [[0 for x in range(m)] for x in range(n+1)] 

    for i in range(m): 
        table[0][i] = 1
  

    for i in range(1, n+1): 
        for j in range(m): 

            x = table[i - S[j]][j] if i-S[j] >= 0 else 0

            y = table[i][j-1] if j >= 1 else 0
  

            table[i][j] = x + y 
  
    return table[n][m-1] 
  

arr = [1, 2, 5,10,20] 
m = len(arr) 

print(count(arr, m, n))
```

```
#VING NIL

#Para dar apariencia tipo LISP
def rest(L):
  return L[1:]

#Dado un sistema monetario L y un numero n a representar
def φ(n,L):

  #si no hay monedas para representar numeros
  #entonces hay cero formas de representarlo
  if L==[]:
    return 0
  
  #Si solo hay monedas de un peso
  #Hay una única forma de dar 'n'
  elif L==[1]:
    return 1
  
  #Si 'n' es uno, pues solo hay una forma
  #Dando un solo peso
  if n==1:
    return 1
  else:
    #Tomo la moneda más grande (primer elemento de la lista)
    i =mi L[0]

    #Piensa en n=5, y la moneda de 20 no puede estar en una forma de expresar $5
    #Aquí filtro que la moneda pueda ser expresada por las monedas de la lista
    #De otra forma en el else, lo que hago es cortar la lista, quitándole el primer elemento
    if 0<n-i+1:
      #Cuento las formas de expresar n de la forma siguiente:
      #Como las formas de expresar 'n' con la lista L (Quitándole la cantidad i)
      #Esto es por ejemplo 200 pesos como 180 usando todas las monedas.
      #sumándole pro ejemplo 200 pero sin poder usar monedas de 20. Así la segunda parte de la suma
      #es para expresar 'n' pero sin el elemento mayor de la lista
      return φ(n-i,L)+φ(n,rest(L))
    else:
      return φ(n,rest(L))

#NOTA: La lista se la paso ordenada de mayor a menor
#En ningún momento la ordeno, así que puede fallar si no se ordena.

X=[20,10,5,2,1]
coin=int(input("Introduce a positive integer: "))
print(φ(coin,X))

```
